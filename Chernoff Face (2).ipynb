```python
# import the required packages
from ChernoffFace import *
import numpy
import matplotlib.cm
import pandas as pd
import numpy as np
```

```python
# import the data
df=pd.read_csv("students_placement_data1.csv");
df.head()
```

```python
# Convert placement status and Gender to 0's and 1's
df['placement status'].replace(['Placed', 'Not placed'],[0, 1], inplace=True)
df['Gender'].replace(['M','F'],[1,0],inplace=True)

# Select only 2 columns
df=df[['placement status','Gender']]
df.head()
```

```python
# for simplicity select first 16 rows
df_16=df.head(16)
df_16
```

```python
# Apply the function
fig = chernoff_face(data=df_16, n_columns=4)
```
